## Find Top 3 by Profit

### Subtask:

## Filter Records by Sales (< 50)

### Subtask:
Filter the `df` DataFrame to show records where 'Sales' are less than 50, similar to `collection.delete_many({'Sales': {'$lt': 50}})`. This will be done by filtering and displaying the records, not actually deleting them.

**Reasoning**:
To filter the `df` DataFrame to show records where 'Sales' are less than 50, I will use boolean indexing on the 'Sales' column. I will then print the number of filtered records and display the first 5 records to verify the operation.



````python
low_sales_df = df[df['Sales'] < 50]
print(f"Number of records with Sales < 50: {len(low_sales_df)}")
print("First 5 records with Sales < 50:")
print(low_sales_df.head())
````

## Aggregate Total Sales by Region

### Subtask:
Group the `df` DataFrame by 'Region' and calculate the sum of 'Sales' for each region, similar to `collection.aggregate([{ '$group': {'_id': '$Region', 'Total_Sales': {'$sum': '$Sales'}} }])`.

**Reasoning**:
To group the `df` DataFrame by 'Region' and calculate the sum of 'Sales' for each region, I will use the `.groupby()` method on the 'Region' column followed by `.sum()` on the 'Sales' column.



````python
total_sales_by_region = df.groupby('Region')['Sales'].sum().reset_index()
print("Total Sales by Region:")
print(total_sales_by_region)
````

## Get Distinct Ship Modes

### Subtask:
Retrieve all unique values from the 'Ship Mode' column in the `df` DataFrame, similar to `distinct('Ship Mode')`.


**Reasoning**:
To retrieve all unique values from the 'Ship Mode' column, I will use the `.unique()` method on that specific column of the `df` DataFrame.

**Reasoning**:
To retrieve and display the distinct ship modes, I will apply the `.unique()` method to the 'Ship Mode' column of the `df` DataFrame and print the result.



````python
distinct_ship_modes = df['Ship Mode'].unique()
print("Distinct Ship Modes:", distinct_ship_modes)
````

**Reasoning**:
To group the `df` DataFrame by 'Category' and count the number of orders in each category, I will use the `.groupby()` method on the 'Category' column followed by `.size()` to count the records in each group. `reset_index()` will convert the result back to a DataFrame with a proper column name for the count.



````python
total_orders_by_category = df.groupby('Category').size().reset_index(name='Total_Orders')
print("Total Orders by Category:")
print(total_orders_by_category)
````
